#### Problema 3
Resta simpleEnunciado: Un número al que se le resta siete da como resultado 18. ¿Cuál es el número?

#### Problema 4:
 Triple con sumaEnunciado: El triple de un número más 4 es igual a 31. ¿Cuál será ese número?Paso 1:
 * Análisis y Solución MatemáticaPlanteamiento: El triple de un número es $3x$. "Más 4" es $+ 4$.Ecuación:$$3x + 4 = 31$$
 * Despeje:
* Primero movemos el 4 (pasa restando): $3x = 31 - 4 \rightarrow 3x = 27$.Luego el 3 que multiplica a la $x$ pasa dividiendo: $x = \frac{27}{3}$.     
* Solución: $x = 9$

```python
# Problema 2: 3x + 4 = 31
# total = 31
# suma = 4
# multiplo = 3
total = int(input("Ingrese el total: "))
suma = int(input("Ingrese la suma: "))
multiplo = int(input("Ingrese el multiplo: "))

# Primero restamos el valor que suma
paso_uno = total - suma
# Luego dividimos por el factor multiplicador
x = paso_uno / multiplo

print(f"El número es: {int(x)}")
print("Programa realizado por Pedro Martinez NL = 34")
```

#### Problema 5:
Geometría (El Terreno)Enunciado: Un terreno tiene un largo que mide 5 metros más que el ancho. Si el perímetro total es 50 m, ¿cuánto mide cada lado?
* Paso 1: Análisis y Solución MatemáticaPlanteamiento: * Ancho ($a$): $x$Largo ($l$): $x + 5$Ecuación: El perímetro de un rectángulo es la suma de sus cuatro lados ($2l + 2a$):$$x + (x + 5) + x + (x + 5) = 50$$Simplificando:$$2(x) + 2(x + 5) = 50$$
* Despeje:Desarrollamos los paréntesis: $2x + 2x + 10 = 50$.
* Sumamos las $x$: $4x + 10 = 50$.Restamos 10: $4x = 50 - 10 \rightarrow 4x = 40$.Dividimos entre 4: $x = \frac{40}{4}$.
* Solución: * Ancho ($x$): $10$ metros.Largo ($x + 5$): $15$ metros.


```python
# Problema 3: 2*(x) + 2*(x + 5) = 50 -> 4x + 10 = 50
# perimetro_total = 50
#diferencia_largo = 5
perimetro_total = int(input("Ingrese el perímetro total: "))
diferencia_largo = int(input("Ingrese la diferencia del largo: "))

# La ecuación simplificada es 4x + (2 * diferencia) = perimetro
# Despejamos: 4x = perimetro - (2 * diferencia)
residuo = perimetro_total - (2 * diferencia_largo)
ancho = residuo / 4
largo = ancho + diferencia_largo

print(f"Las medidas del terreno son: Ancho = {ancho}m, Largo = {largo}m")
print("Programa realizado por Pedro Martinez NL = 34")
```
